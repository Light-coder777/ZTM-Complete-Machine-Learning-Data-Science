## Import Data

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
col_names = ["Alcohol", "Malic Acid", "Ash", "Alcalinity of Ash", "Magnesium", "Total Phenols", "Flavanoids", "Nonflavnoid Phenols", "Proanthocyanins", "Color Intensity", "Hue", "OD280/OD315", "Proline"]
wine_df = pd.read_csv("../data/wine_dataset/wine.data",
                      names=col_names)
wine_df[:10]
```

```python
# check for missing data
null_columns = wine_df.columns[wine_df.isnull().any()]
wine_df[null_columns].isnull().sum()
```

## Plot Data

```python
plt.style.use("seaborn-whitegrid")
```

```python
fig, ((ax0, ax1)) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,8))

fig.suptitle("Wine Data Analysis", fontsize=16, fontweight="bold")

# first plot
scatter = ax0.scatter(wine_df["Alcohol"],
                      wine_df["Hue"],
                      c=wine_df.index,
                      cmap="spring")
ax0.set(title="Alcohol and Hue Levels",
        xlabel="Alcohol",
        ylabel="Hue")
ax0.legend(*scatter.legend_elements(), title="Class Identifier")
ax0.axhline(y=wine_df["Hue"].mean(),
            color="b",
            linestyle="--",
            label="Average Hue")
ax0.set_xlim([11,15])

# second plot
scatter = ax1.scatter(wine_df["Malic Acid"],
                      wine_df["Alcalinity of Ash"],
                      c=wine_df.index,
                      cmap="spring")
ax1.set(title="Malic Acid and Alcalinity of Ash",
        xlabel="Malic Acid",
        ylabel="Alcalinity of Ash")
ax1.axhline(y=wine_df["Alcalinity of Ash"].mean(),
            color="b",
            linestyle="--",
            label="Average Alcalinity of Ash");
ax1.set_xlim([0,6])
ax1.set_ylim([10,31])

# adjust space between subplots
plt.subplots_adjust(hspace=0.3);
```
